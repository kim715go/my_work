```python
titles = titles.apply(lambda x : "".join(c for c in x if c not in string.punctuation))
```

```python
import re
```

```python
nons = re.compile(r'[^가-힣a-zA-Z0-9\s]')

titles = titles.apply(lambda x : re.sub(nons, "", x))
```

```python
frequency = {}

for line in titles : 
    temp = line.strip().split()
    for item in temp : 
        if item not in frequency : 
            frequency[item] = 1
        frequency[item] += 1
```

```python
frequencySorted = sorted(frequency.items(), key=lambda x : x[1], reverse=True)
```
