```python
import requests
```

```python
defaultURL = "https://dapi.kakao.com/v2/local/search/address"
headers = {'Content-Type': 'application/json; charset=utf-8',
                'Authorization': 'KakaoAK 8e0e3db91e6957e61929511be2e3787d'}
oneResOutput = 15
parameters = {'sort':'accuracy', 'page':1, 'size':oneResOutput, 'query':""} #size : the num of result for each page!
```

```python
parameters['query'] = "은천로5길 35"
```

```python
res = requests.get(defaultURL, headers = headers, params = parameters)
```

```python
res
```

```python
res.json()['documents'][0]
```

```python
for place in res.json()['documents'] : 
    print("\t".join([place['place_name'], place['address_name'], place['x'], place['y'], place['place_url']]))
```

```python
import pandas as pd
```

```python
apartments = pd.read_excel('추가 아파트.xlsx', sheet_name=1)
```

```python
apartments
```

```python
searchSession = requests.Session()
```

```python
res = searchSession.get(defaultURL, headers=headers, params=parameters)
```

```python
res.json()['documents'][0]
```

```python
for apart in apartments.iterrows():
    parameters['query'] = apart[1].검색어
    tempRes = searchSession.get(defaultURL, headers=headers, params = parameters)
    try : 
        firstItem = tempRes.json()['documents'][0]
    except IndexError : 
        print(apart[1].검색어)
        continue
    else:
        apartments.loc[apart[0], 'x'] = firstItem['x']
        apartments.loc[apart[0], 'y'] = firstItem['y']
        apartments.loc[apart[0], '검색 주소'] = firstItem['address_name']
```

```python
apartments.to_excel('추가 검색결과.xlsx')
```
